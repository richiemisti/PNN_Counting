```python
import os
import glob
import pandas as pd
from matplotlib import cm
cmap = cm.get_cmap('viridis')
from rolling_ball import rolling_ball_radius
from draw_and_measure import draw_and_measure
```

```python
parent_dir = 'D:\PNN_Counting'
```

```python
subfolders = [ f.path for f in os.scandir(parent_dir) if f.is_dir() ] # lists all the folders inside the parent directory

# for the purposes here, I'm only going to focus on the WFA and the PV mice folders - the Analysis folder has the colocalization information

master_df = pd.DataFrame() # master averages concatenated
raw_df = pd.DataFrame() # all raw data concatenated

for subfolder in subfolders:
    if 'Mice' in subfolder:
        #print(subfolder) # it will do all the WFA ones first, for example and then all the PV 
        new_subfolders = [ f.path for f in os.scandir(subfolder) if f.is_dir() ] # this should give me a list of all the mice in here
        # next I need to go through the list of all the mice and get into their predictions folder
        #print(new_subfolders)

        for mouse_folder in range(len(new_subfolders)):
            my_folder = [ f.path for f in os.scandir(new_subfolders[mouse_folder]) if f.is_dir() ]

            #print(my_folder)
            #print('next')

            for prediction_folder in my_folder:
                print(f"Section prediction folder: {prediction_folder}")
                csv_file = glob.glob(os.path.join(prediction_folder,"*.csv"))[0]
                tif_file = glob.glob(os.path.join(prediction_folder,"*.tif"))[0]
                #print(csv_file)

                img_bg_subtracted = rolling_ball_radius(tif_file,radius=22)

                df, df_master, output_image, intensity_means, intesnity_stdev, intensity_sterr, combined_pnn_mask = draw_and_measure(img_bg_subtracted, csv_file, diameter=40, cmap=cmap)

                
                master_df = pd.concat([master_df,df_master],ignore_index=True)
                raw_df = pd.concat([raw_df,df],ignore_index=True)

# save the df files and then enjoy life :)
master_df.to_csv('master_df.csv', index=False)
raw_df.to_csv('raw_df.csv', index=False)

        
```
